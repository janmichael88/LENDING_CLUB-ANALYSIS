```python
#grab dependencies
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import os
from functools import partial,reduce
import seaborn as sns

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
#get path
path = os.getcwd()
#get list of files
loan_files = os.listdir(path)
#get only county files
loan_files = [file for file in loan_files if ".csv" in file]
```

```python
loan_files
#notes "issue_d" should be preserved
```

```python
#the real thing
frames_first_10 = []
for i in range(len(loan_files)):
    #load in frame
    frame = pd.read_csv(loan_files[i],header=1)
    #sort column index alphabetically
    frame = frame.sort_index(axis=1)
    #get list of column frame names
    list_col_names = frame.columns.tolist()
    #remove 'id','member_id','issue_d' from list, these will be added to the front
    column_names_to_remove = ['issue_d','id','member_id']
    for name in column_names_to_remove:
        list_col_names.remove(name)
    #bring back column_names_to_remove to the front
    new_column_ordering = column_names_to_remove + list_col_names
    frame = frame[new_column_ordering]
    #append frame to list
    frames_first_10.append(frame)

final_stacked_frame = pd.concat(frames_first_10)
final_stacked_frame = final_stacked_frame.reset_index()
```

```python
#checking date columns to NA'a
sum(final_stacked_frame['issue_d'].isnull())
#there are 25 rows missing issue d value
```

```python
#finds rows that have missing issue_d
final_stacked_frame[final_stacked_frame['issue_d'].isnull()]
```

```python
#check same number of columns across files
for frame in frames_first_10:
    print(len(frame.columns))
    
#all have the same number of columns
#check if column names are the same among the data frames
#list of sorted columns for each from
list_col_names_sorted_frame = []
for frame in frames_first_10:
    #get sorted list of columnes
    list_col_sorted = frame.sort_index(axis=1).columns
    #added list_col_sorted to list_col_names_sorted_frame
    list_col_names_sorted_frame.append(list_col_sorted)

#check unique column names
len(np.unique(list_col_names_sorted_frame))
#should return 1 list of length 144
#dataframes match
```
